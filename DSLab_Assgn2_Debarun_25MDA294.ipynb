## Q1. Write a Python program that prompts the user for his/her price of a meal in a restaurant, then reports how much GST she has to be paid and total bill amount.
Prompts the user for meal price, reports GST and total bill.

```python
print("=== GST Calculator ===")
meal_price = float(input("Enter meal price (₹): "))
gst_rate = 0.18
gst = meal_price * gst_rate
total = meal_price + gst
print(f"\nMeal Price : ₹{meal_price:.2f}")
print(f"GST (18%)  : ₹{gst:.2f}")
print(f"Total Bill : ₹{total:.2f}")
```

## Q2. Evaluate the quadratic equation for a given a, b, and c.
Evaluate the quadratic equation for user-given a, b, and c.

```python
import math

print("=== Quadratic Equation Solver (ax² + bx + c = 0) ===")
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

discriminant = b**2 - 4*a*c
print(f"\nDiscriminant = {discriminant}")

if discriminant > 0:
    x1 = (-b + math.sqrt(discriminant)) / (2*a)
    x2 = (-b - math.sqrt(discriminant)) / (2*a)
    print(f"Two real roots: x1 = {x1:.2f}, x2 = {x2:.2f}")
elif discriminant == 0:
    x = -b / (2*a)
    print(f"One real root: x = {x:.2f}")
else:
    real = -b / (2*a)
    imag = math.sqrt(-discriminant) / (2*a)
    print(f"Complex roots: x1 = {real:.2f}+{imag:.2f}i, x2 = {real:.2f}-{imag:.2f}i")
```

## Q3. How would we print the "99 Bottles of Beer" song?
User chooses how many verses to print.

## Q4. Write a Python program that computes the factorial of an integer.
Computes the factorial of a user-entered integer.

```python
print("=== Factorial Calculator ===")
n = int(input("Enter a non-negative integer: "))

if n < 0:
    print("Factorial is not defined for negative numbers.")
else:
    result = 1
    for i in range(2, n + 1):
        result *= i
    print(f"\n{n}! = {result}")
```

## Q5. Write code to display and count the factors of a number.
Displays and counts all factors of a user-entered number.

```python
print("=== Factor Finder ===")
n = int(input("Enter a positive integer: "))

factors = [i for i in range(1, n + 1) if n % i == 0]

print(f"\nFactors of {n}: {factors}")
print(f"Total count : {len(factors)}")
```

## Q6. Create a 2D Numpy array and extract a subarray
User enters array size; program creates array and extracts a subarray.

```python
import numpy as np

print("=== 2D NumPy Array & Subarray ===")
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of cols: "))

arr = np.arange(1, rows * cols + 1).reshape(rows, cols)
print("\nOriginal Array:")
print(arr)

print("\nEnter subarray range (0-indexed):")
r1 = int(input("  Row start   : "))
r2 = int(input("  Row end     : "))
c1 = int(input("  Col start   : "))
c2 = int(input("  Col end     : "))

subarray = arr[r1:r2, c1:c2]
print(f"\nSubarray arr[{r1}:{r2}, {c1}:{c2}]:")
print(subarray)
```

## Q7. Modify elements using slicing and broadcasting.
User picks a row to zero-out and a column to add a value to.

```python
import numpy as np

print("=== Slicing & Broadcasting ===")
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of cols: "))

arr = np.arange(1, rows * cols + 1).reshape(rows, cols)
print("\nOriginal Array:")
print(arr)

row_idx = int(input(f"\nEnter row index to set to 0 (0 to {rows-1}): "))
arr[row_idx, :] = 0
print(f"After setting row {row_idx} to 0 (slicing):")
print(arr)

col_idx = int(input(f"\nEnter col index to broadcast add (0 to {cols-1}): "))
val = int(input("Enter value to add: "))
arr[:, col_idx] += val
print(f"After adding {val} to column {col_idx}:")
print(arr)
```

## Q8. Create a dataframe with name, age and marks.
User enters student records; program builds and displays a DataFrame.

```python
import pandas as pd

print("=== Create Student DataFrame ===")
n = int(input("How many students to enter? "))

names, ages, marks = [], [], []
for i in range(n):
    print(f"\n--- Student {i+1} ---")
    names.append(input("  Name  : "))
    ages.append(int(input("  Age   : ")))
    marks.append(float(input("  Marks : ")))

df = pd.DataFrame({'name': names, 'age': ages, 'marks': marks})
print("\nDataFrame:")
print(df)
```

## Q9. Filter students with marks > 80, sort by age.

```python
import pandas as pd

print("=== Filter & Sort DataFrame ===")
n = int(input("How many students to enter? "))

names, ages, marks = [], [], []
for i in range(n):
    print(f"\n--- Student {i+1} ---")
    names.append(input("  Name  : "))
    ages.append(int(input("  Age   : ")))
    marks.append(float(input("  Marks : ")))

df = pd.DataFrame({'name': names, 'age': ages, 'marks': marks})

threshold = float(input("\nEnter marks threshold to filter (e.g. 80): "))
filtered = df[df['marks'] > threshold].sort_values('age').reset_index(drop=True)

print(f"\nStudents with marks > {threshold}, sorted by age:")
print(filtered)
```

## Q10. Export the filtered Dataframe to csv.

```python
import pandas as pd

print("=== Export Filtered DataFrame to CSV ===")
n = int(input("How many students to enter? "))

names, ages, marks = [], [], []
for i in range(n):
    print(f"\n--- Student {i+1} ---")
    names.append(input("  Name  : "))
    ages.append(int(input("  Age   : ")))
    marks.append(float(input("  Marks : ")))

df = pd.DataFrame({'name': names, 'age': ages, 'marks': marks})

threshold = float(input("\nEnter marks threshold to filter (e.g. 80): "))
filtered = df[df['marks'] > threshold].sort_values('age').reset_index(drop=True)

filename = input("Enter output CSV filename (e.g. result.csv): ").strip()
if not filename.endswith('.csv'):
    filename += '.csv'

filtered.to_csv(filename, index=False)
print(f"\nExported {len(filtered)} record(s) to '{filename}'")
print(filtered)
```
