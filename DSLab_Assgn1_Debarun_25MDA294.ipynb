# 1. Conditional Logic & Control Flow

**Requirements:**
- Check if number is prime
- Print all factors if not prime
- Classify as even/odd and positive/negative

```python
def check_prime_and_factors(n):
    is_prime = True
    factors = []
    
    if n > 1:
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                is_prime = False
                break
        for i in range(1, n + 1):
            if n % i == 0:
                factors.append(i)
    
    parity = 'Even' if n % 2 == 0 else 'Odd'
    sign = 'Positive' if n > 0 else ('Negative' if n < 0 else 'Zero')
    
    return {'is_prime': is_prime, 'factors': factors, 'parity': parity, 'sign': sign}

test_numbers = [17, 25, 20, 1, -5, 0]

for num in test_numbers:
    result = check_prime_and_factors(num)
    print(f"Number: {num} | Prime: {result['is_prime']} | Factors: {result['factors']} | {result['parity']}, {result['sign']}")
```

# 2. Loops & Pattern Generation

**Requirements:**
- Generate pattern for n=5: 1, 12, 123, 1234, 12345
- Write logic to reverse the pattern

```python
def forward_pattern(n):
    for i in range(1, n + 1):
        print(''.join(str(j) for j in range(1, i + 1)))

def reverse_pattern(n):
    for i in range(n, 0, -1):
        print(''.join(str(j) for j in range(i, 0, -1)))

print("Forward Pattern:")
forward_pattern(5)

print("\nReverse Pattern:")
reverse_pattern(5)
```

# 3. Functions & Recursion

## 3a. Fibonacci Series

**Requirement:** Write a recursive function to compute Fibonacci series up to n terms

```python
def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n-1) + fib_recursive(n-2)

n = 10
print("Recursive:", [fib_recursive(i) for i in range(n)])
```

## 3b & 3c. Factorial - Recursion vs Iteration & Efficiency Comparison

**Requirements:**
- Write function to compute factorial using recursion
- Write function to compute factorial using iteration
- Compare both approaches in terms of efficiency (execution time & complexity)

```python
def factorial_rec(n):
    return 1 if n <= 1 else n * factorial_rec(n - 1)

def factorial_iter(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

print("3b & 3c. Factorial - Recursion vs Iteration\n")
print(f"{'N':<5} {'Recursive (μs)':<18} {'Iterative (μs)':<18} {'Speedup':<10}")
print("-" * 55)

iterations = 10000
speedups = []
for n in [10, 15, 20, 25]:
    t1 = time.time()
    for _ in range(iterations):
        factorial_rec(n)
    t1 = ((time.time() - t1) / iterations) * 1e6
    
    t2 = time.time()
    for _ in range(iterations):
        factorial_iter(n)
    t2 = ((time.time() - t2) / iterations) * 1e6
    
    speedup = t1 / t2 if t2 > 0 else float('inf')
    speedups.append(speedup)
    print(f"{n:<5} {t1:<18.2f} {t2:<18.2f} {speedup:.2f}x")

avg_speedup = sum(speedups) / len(speedups)

print("\n" + "="*55)
print("CONCLUSION:")
print(f"Iterative is faster (~{avg_speedup:.1f}x average speedup)")
print("Reason: Function call overhead in recursion")
print("Time Complexity: O(n) | Space: O(n) recursive vs O(1) iterative")
```

# 4. Lists & Dictionaries

**Requirements:**
- Remove duplicates without using set()
- Count frequency of each element using dictionary
- Find second largest element in list

```python
def remove_duplicates(lst):
    result = []
    for x in lst:
        if x not in result:
            result.append(x)
    return result

def count_freq(lst):
    freq = {}
    for x in lst:
        freq[x] = freq.get(x, 0) + 1
    return freq

def second_largest(lst):
    largest = max(lst)
    return max(x for x in lst if x != largest)

lst = [1, 2, 3, 2, 4, 1, 5, 3, 2, 6, 5]
print("Original:", lst)
print("4a. Unique:", remove_duplicates(lst))
print("4b. Frequency:", count_freq(lst))
print("4c. Second Largest:", second_largest(lst))
```

# 5. Problem Solving (Mini Challenge)

## Marks = [78, 85, 62, 90, 55, 73, 88, 91, 67, 80]

**Requirements:**
- a) Categorize students into grades (A: 90-100, B: 80-89, C: 70-79, D: 60-69, F: <60)
- b) Find top 3 students
- c) Display results in structured format
- d) Extend logic to handle dynamic input size

```python
def get_grade(marks):
    if marks >= 90: return 'A'
    elif marks >= 80: return 'B'
    elif marks >= 70: return 'C'
    elif marks >= 60: return 'D'
    else: return 'F'

marks = [78, 85, 62, 90, 55, 73, 88, 91, 67, 80]
students = [(i+1, m, get_grade(m)) for i, m in enumerate(marks)]

print("5a. Student Grades:")
print(f"{'ID':<5} {'Marks':<8} {'Grade':<6}")
for sid, m, g in students:
    print(f"{sid:<5} {m:<8} {g:<6}")

print("\n5b.Top 3 Students:")
top3 = sorted(students, key=lambda x: x[1], reverse=True)[:3]
for rank, (sid, m, g) in enumerate(top3, 1):
    print(f"{rank}. Student {sid}: {m} marks (Grade {g})")

print("\n5c. Summary Statistics:")
print(f"Average: {sum(m for _, m, _ in students)/len(marks):.2f}")
print(f"Highest: {max(m for _, m, _ in students)}")
print(f"Lowest: {min(m for _, m, _ in students)}")
```

## 5d. Dynamic Input Handling

**Dynamic processing with different marking scales**

```python
def process_marks(marks_list, scale=100):
    data = {
        'Student_ID': list(range(1, len(marks_list) + 1)),
        'Marks': marks_list,
        'Grade': [get_grade(m) for m in marks_list],
        'Percentage': [round((m/scale)*100, 1) for m in marks_list]
    }
    df = pd.DataFrame(data)
    return df.sort_values('Marks', ascending=False)

print("Scale 100:")
df1 = process_marks([78, 85, 62, 90, 55, 73, 88, 91, 67, 80], 100)
print(df1)

print("\n\nScale 50:")
df2 = process_marks([39, 42, 31, 45, 27, 36, 44, 45, 33, 40], 50)
print(df2)
```
