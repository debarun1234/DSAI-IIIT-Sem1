## Example 1 — Coefficient of Variation

Suppose we have a series with **mean of 35** and a **variance of 80**. Find out the **Coefficient of Variation (CV)**.

> **Formula:**  
> σ = √Variance  
> CV = (σ / Mean) × 100

````python
import math

# Given values
mean     = 35
variance = 80

# Step 1: Standard Deviation = square root of Variance
std_dev = math.sqrt(variance)
print(f"Standard Deviation (σ) = √{variance} = {std_dev:.4f}")

# Step 2: Coefficient of Variation
cv = (std_dev / mean) * 100
print(f"\nCoefficient of Variation (CV) = (σ / Mean) × 100")
print(f"CV = ({std_dev:.4f} / {mean}) × 100")
print(f"CV = {cv:.2f}%")
````

### Analysis — Example 1

| Parameter | Value |
|---|---|
| Mean (x̄) | 35 |
| Variance (σ²) | 80 |
| Standard Deviation (σ) | √80 ≈ 8.9443 |
| **Coefficient of Variation (CV)** | **≈ 25.55%** |

**Interpretation:**  
A CV of ~25.55% indicates moderate relative variability. The standard deviation is about 25.55% of the mean, meaning the data is reasonably spread around the mean but not excessively scattered.

## Example 2 — Finding Standard Deviation from CV

Given two series:
- **Series 1**: CV = 80, Mean = 25
- **Series 2**: CV = 90, Mean = 35

Find the **Standard Deviation** for both series.

> **Formula:**  
> CV = (σ / Mean) × 100  
> ⟹  σ = (CV × Mean) / 100

````python
# Given values
cv1, mean1 = 80, 25
cv2, mean2 = 90, 35

# Standard Deviation = (CV × Mean) / 100
sd1 = (cv1 * mean1) / 100
sd2 = (cv2 * mean2) / 100

print("Series 1:")
print(f"  CV = {cv1}, Mean = {mean1}")
print(f"  σ₁ = ({cv1} × {mean1}) / 100 = {sd1}")

print("\nSeries 2:")
print(f"  CV = {cv2}, Mean = {mean2}")
print(f"  σ₂ = ({cv2} × {mean2}) / 100 = {sd2}")

print("\n--- Summary ---")
print(f"Standard Deviation of Series 1 (σ₁) = {sd1}")
print(f"Standard Deviation of Series 2 (σ₂) = {sd2}")
````

### Analysis — Example 2

| Series | CV | Mean | Standard Deviation (σ) |
|---|---|---|---|
| Series 1 | 80 | 25 | **20** |
| Series 2 | 90 | 35 | **31.5** |

**Interpretation:**  
- Series 2 has a higher CV (90 vs 80) meaning it is **relatively more variable** than Series 1.
- Series 2 also has a higher absolute standard deviation (31.5 vs 20) — its values are spread farther from its mean in absolute terms.
- Despite Series 2 having a larger mean, CV captures the *relative* spread, confirming it is the more dispersed dataset proportionally.

## Example 3 — Frequency Distribution Table

Draw the frequency distribution table for the following data:  
`2, 3, 1, 4, 2, 2, 3, 1, 4, 4, 4, 2, 2, 2`

````python
import pandas as pd
from collections import Counter

# Raw data
data = [2, 3, 1, 4, 2, 2, 3, 1, 4, 4, 4, 2, 2, 2]

# Count frequencies
freq_count = Counter(data)

# Build frequency distribution table
values     = sorted(freq_count.keys())
freq       = [freq_count[v] for v in values]
rel_freq   = [round(f / len(data) * 100, 2) for f in freq]

df_freq = pd.DataFrame({
    'Value (x)':           values,
    'Frequency (f)':       freq,
    'Relative Frequency (%)': rel_freq
})

# Add totals row
totals = pd.DataFrame({
    'Value (x)':              ['Total'],
    'Frequency (f)':          [sum(freq)],
    'Relative Frequency (%)': [sum(rel_freq)]
})
df_freq = pd.concat([df_freq, totals], ignore_index=True)

print("Frequency Distribution Table")
print("=" * 45)
print(df_freq.to_string(index=False))

print(f"\nTotal observations : {len(data)}")
print(f"Mode               : {max(freq_count, key=freq_count.get)} (appears {max(freq_count.values())} times)")
````

### Analysis — Example 3

| Value (x) | Frequency (f) | Relative Frequency (%) |
|---|---|---|
| 1 | 2 | 14.29% |
| 2 | 6 | 42.86% |
| 3 | 2 | 14.29% |
| 4 | 4 | 28.57% |
| **Total** | **14** | **100%** |

**Interpretation:**  
- The dataset has **14 observations** with values ranging from 1 to 4.
- **Value 2 is the mode** — it appears 6 times (≈ 43% of data), making it the most frequent value.
- Values 1 and 3 appear equally (2 times each), while value 4 appears 4 times.
- The distribution is **not uniform** — value 2 dominates the dataset.

## Example 4 — Less-than Cumulative Frequency Distribution

The table below gives the values of temperature (°C) recorded in Bangalore for 25 days in summer.  
Represent the data in the form of a **less-than-type cumulative frequency distribution**.

```
22  23  21  27  20
25  20  24  23  27
37  25  34  32  26
29  29  38  34  29
23  31  20  29  22
```

````python
import pandas as pd
import matplotlib.pyplot as plt

# Raw temperature data (25 values)
temps = [
    22, 23, 21, 27, 20,
    25, 20, 24, 23, 27,
    37, 25, 34, 32, 26,
    29, 29, 38, 34, 29,
    23, 31, 20, 29, 22
]

print(f"Total observations : {len(temps)}")
print(f"Min temperature    : {min(temps)}°C")
print(f"Max temperature    : {max(temps)}°C")
print(f"Sorted data        : {sorted(temps)}")

# Define class intervals (width = 4, from 20 to 40)
bins       = [20, 24, 28, 32, 36, 40]
labels     = ['20 – 24', '24 – 28', '28 – 32', '32 – 36', '36 – 40']
upper_lims = ['Less than 24', 'Less than 28', 'Less than 32',
              'Less than 36', 'Less than 40']

# Count frequencies per interval (right=False: left-inclusive, right-exclusive)
freq = []
for i in range(len(bins) - 1):
    count = sum(bins[i] <= t < bins[i+1] for t in temps)
    freq.append(count)

# Cumulative frequency
cum_freq = []
running  = 0
for f in freq:
    running += f
    cum_freq.append(running)

# Build the table
df_cf = pd.DataFrame({
    'Class Interval (°C)':        labels,
    'Frequency (f)':              freq,
    'Upper Class Limit':          upper_lims,
    'Cumulative Frequency':       cum_freq
})

totals_row = pd.DataFrame({
    'Class Interval (°C)': ['Total'],
    'Frequency (f)':       [sum(freq)],
    'Upper Class Limit':   [''],
    'Cumulative Frequency':[''],
})
df_cf = pd.concat([df_cf, totals_row], ignore_index=True)

print("\nLess-than Cumulative Frequency Distribution")
print("=" * 65)
print(df_cf.to_string(index=False))

# --- Plot the Ogive (Less-than Cumulative Frequency Curve) ---
upper_bounds = [24, 28, 32, 36, 40]

plt.figure(figsize=(8, 5))
plt.plot([20] + upper_bounds, [0] + cum_freq,
         marker='o', color='steelblue', linewidth=2, markersize=7, label='Less-than Ogive')
plt.fill_between([20] + upper_bounds, [0] + cum_freq, alpha=0.15, color='steelblue')

for x, y in zip(upper_bounds, cum_freq):
    plt.annotate(f'{y}', xy=(x, y), xytext=(x + 0.3, y + 0.5),
                 fontsize=10, color='navy')

plt.xticks(upper_bounds, [f'< {b}' for b in upper_bounds])
plt.xlabel('Temperature (°C) — Upper Class Boundary', fontsize=11)
plt.ylabel('Cumulative Frequency', fontsize=11)
plt.title('Less-than Ogive — Bangalore Summer Temperatures (25 days)', fontsize=12)
plt.yticks(range(0, 27, 5))
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
````

### Analysis — Example 4

| Class Interval (°C) | Frequency (f) | Upper Class Limit | Cumulative Frequency |
|---|---|---|---|
| 20 – 24 | 9 | Less than 24 | 9 |
| 24 – 28 | 6 | Less than 28 | 15 |
| 28 – 32 | 5 | Less than 32 | 20 |
| 32 – 36 | 3 | Less than 36 | 23 |
| 36 – 40 | 2 | Less than 40 | 25 |
| **Total** | **25** | | |

**Interpretation:**  
- The temperature range is **20°C to 38°C**, divided into 5 equal class intervals of width 4.
- **9 out of 25 days** (36%) had temperatures below 24°C — the most frequent class, indicating the majority of days were cooler.
- **20 out of 25 days** (80%) had temperatures below 32°C, showing that extreme heat (32°C+) was relatively rare.
- The **Ogive (less-than curve)** rises steeply at lower temperatures and flattens at higher temperatures, confirming a **right-skewed distribution** — most days were in the 20–28°C range.
- The cumulative frequency reaches **25 at 'Less than 40'**, which accounts for all observations, confirming data completeness.
