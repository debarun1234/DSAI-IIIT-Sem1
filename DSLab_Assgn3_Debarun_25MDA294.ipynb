**Q2**. Prime number are the unique set of numbers widely used in 
mathematics. It is  because of the fact that the number is 
divisible by 1 and itself. Now construct a function that 
determines the whether the given is prime or not and further 
reuse this function to print all the prime number between 1 
and the N(like if N=10 prime numbers are 2,3,5,7 )

````python
#solve q2
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def print_primes(n):
    for i in range(2, n + 1):
        if is_prime(i):
            print(i)
# Example usage:
N = 10
print(f"Prime numbers between 1 and {N}:")
print_primes(N)
````

**Q10**. You are given a dataset containing student information in a raw and inconsistent format. Your task is to clean the data and perform basic analysis.

```python
raw_data = [
    "Rahul,78,20",
    "Anita,85,21",
    "John,NA,19",
    "Meena,92,20",
    "Kiran,,22",
    "Arjun,67,twenty",
    "Divya,88,21"
]
```

Tasks:
1. **Data Cleaning** — remove invalid records (missing values, 'NA', non-numeric marks or age); convert marks and age to int
2. Calculate **Average Marks** after cleaning
3. Display names of students whose **marks are even**
4. Display the name of the **topper**
5. Allocate **batches** (size ≥ 2) and display

````python
raw_data = [
    "Rahul,78,20",
    "Anita,85,21",
    "John,NA,19",
    "Meena,92,20",
    "Kiran,,22",
    "Arjun,67,twenty",
    "Divya,88,21"
]

# 1. Data Cleaning
cleaned = []
for entry in raw_data:
    entry = entry.strip().strip('"').strip(',')
    parts = entry.split(',')
    if len(parts) != 3:
        continue
    name, marks, age = parts
    try:
        marks = int(marks)
        age   = int(age)
        cleaned.append({'name': name, 'marks': marks, 'age': age})
    except ValueError:
        pass   # skip records with NA or non-numeric values

print('Cleaned data:')
for s in cleaned:
    print(f"  {s['name']}: marks={s['marks']}, age={s['age']}")

# 2. Average Marks
avg = sum(s['marks'] for s in cleaned) / len(cleaned)
print(f'\nAverage Marks: {avg:.2f}')

# 3. Students with even marks
print('\nStudents with even marks:')
for s in cleaned:
    if s['marks'] % 2 == 0:
        print(f"  {s['name']} ({s['marks']})")

# 4. Topper
topper = max(cleaned, key=lambda s: s['marks'])
print(f"\nTopper: {topper['name']} ({topper['marks']})")

# 5. Batch allocation (batch size = 2)
batch_size = 2
print('\nBatch Allocation:')
for i, s in enumerate(cleaned):
    batch_num = i // batch_size
    print(f"  {s['name']} → Batch {batch_num}")
````

**Output** of Q10:

1. **Invalid records removed**: John (NA marks), Kiran (missing marks), Arjun (non-numeric age)
1a. **Cleaned students**: Rahul(78), Anita(85), Meena(92), Divya(88)
2. **Average Marks**: (78+85+92+88) / 4 = **85.75**
3. **Even marks**: Rahul (78), Meena (92), Divya (88)
4. **Topper**: Meena (92)
5. **Batches**: Batch 0 → Rahul, Anita | Batch 1 → Meena, Divya
